```python
import re
Input_Text=text
pattern=r'([a-zA-Z0-9\-\.]+@[a-zA-z\.]{2,})'
email_id=re.findall(pattern, Input_Text, flags=re.M|re.I)
    
email_file=[]
for ids in email_id:
    e_ids=ids.split('@')[1]
    #print(e_ids)
    domain=e_ids.split('.')
    domain=' '.join(list(d for d in domain if (len(d)>2 and d!='com')))
    email_file.append(domain)
print(email_file)
```

```python
text='The great thoughts Ballantine Books on New York USA dictionary of quotations of different kind concentrating on statements'

from nltk.chunk import ne_chunk
import nltk

tokens=nltk.word_tokenize(text)
pos=nltk.pos_tag(tokens)
chunks=ne_chunk(pos, binary=False)
list(chunks)
```

```python
person_entities=[]
named_entities=[]

for t in chunks.subtrees():
    #print(t.label(),t)
    if t.label() == 'PERSON':
        person_entities.append(list(t))
        
for entity in person_entities:
    word=[entity[x][0] for x in range(len(entity))]
    pattern=str(' '.join(word))
    text11=re.sub(pattern,'',text)
    

for t in chunks.subtrees():
    if t.label() == 'GPE':
        named_entities.append(list(t))
        
for entity in named_entities:
    word=[entity[x][0] for x in range(len(entity))]
    print(entity)
    pattern=str(' '.join(word))
    text=re.sub(pattern,'_'.join(word),text)
    
print(text.lower())
```
