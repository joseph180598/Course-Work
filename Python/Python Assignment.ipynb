<br>Cosine distance between two points (x,y) and (p,q) is defind as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)

```python
import math as m

def cdis(R,S):
    '''This function calculates the distance based on formula
    '''
    (x,y)=R
    (p,q)=S
    up=((x*p)+(y*q))
    down=((m.sqrt((x**2)+(y**2)))*(m.sqrt((p**2)+(q**2))))
    fin=up/down
    ans=m.acos(fin)
    return ans

def closest_points_to_p(S, P):
    '''This function calculates the distance and sorts them
    '''
    lst=[]
    for x in S:
        n=cdis(x,P)
        tup=(n,x)
        lst.append(tup)
        det=dict(lst) 
    det2=sorted(det)
    for p in range(5):
        print(det[det2[p]])  
    
S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
closest_points_to_p(S,P)
    # write your code here
    #return closest_points_to_p 
```
